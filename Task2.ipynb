```python
# First of all we must import the libraries:

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn import datasets
```

```python
# Let's descove the data:

path = "Iris.csv"
data = pd.read_csv(path)
print("Data imported successfully")
data.head(10)
```

```python
# We don't need id column 
data.drop("Id", axis=1, inplace=True) 
data.head(10)
```

```python
print("Our dataset consist of {} samples".format(data.shape[0]))
print('-'*50)
print("There are {} features: {}".format(data.shape[1],data.columns.values))
print('-'*50)
print(data.describe())
print('-'*50)
print("Missed values:")
data.isnull().sum()
```

There are no missed values

```python
print(data.Species.value_counts())
```

```python
# Let's visualize our data:

sns.pairplot(data, hue= "Species")
```

```python
sns.pairplot(data, hue= "Species" , kind = "kde")
```

 It is very clear that there are three clusters, so **Let's start clustering**

```python
from sklearn.cluster import KMeans
points = data.iloc[: , :4]
kmeans = KMeans(n_clusters=3, random_state=0).fit(points)
sns.scatterplot(data= data , x = "PetalWidthCm" , y = "PetalLengthCm" , hue = "Species" , s=70)
plt.scatter(x = kmeans.cluster_centers_[: , 3] , y = kmeans.cluster_centers_[: , 2] ,c="red" ,s=80)
plt.show()
```

The clusters centroids are colored red
